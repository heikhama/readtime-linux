```python
import os
import re


#function to add new contact

def add_new_contact(contact_data, c_phone, c_name):
  mobileno_pattern = re.compile(r'^\d{10}$')
  if re.match(mobileno_pattern, c_phone):
        print("Mobile Number is correct")
        contact_data[c_name] = c_phone
  else:
        print("Enter again")

#function to delete old contact

def del_old_contact(contact_data, c_name):
  if c_name in contact_data:
    del contact_data[c_name]
    print(f"{c_name}'s contact is deleted.")
  else:
    print(f"{c_name}'s Given Name is not found in Database.")

#function to update old contact

def update_old_contactphone(contact_data, c_name, new_phonenum):
  if c_name in contact_data:
    contact_data[c_name] = new_phonenum
    print(f"{c_name}'s contact information database is updated.")
  else:
    print(f"{c_name} not found in contact Database.")

#function to update old contact

def update_old_contactname(contact_data, c_new_name, c_phone):
  if c_phone in contact_data:
    contact_data[c_phone] = c_new_name
    print(f"{c_phone}'s contact information database is updated.")
  else:
    print(f"{c_phone} not found in contact Database.")

#function to view contact details

def view_all_contactdetails(contact_data):
  print("Contact Management database: \n")
  for c_name, c_phone in contact_data.items():
    print(f"{c_name}: {c_phone} \n")

#Sorting the list of the dictionary
  sorted_dict = dict(sorted(contact_data.items(), key=lambda item: item[1]))
  print(sorted_dict)

#function to save the contact data in TXT file

def save_contact_file(contact_data, filename):
  with open(filename, 'w') as file:
    for c_name, c_phone in contact_data.items():
      file.write(f"{c_name}, {c_phone} \n")

#function to load contacts from File

def load_contactdetails(filename):
  contact_data={}
  if os.path.exists(filename):
    with open(filename, 'r') as file:
      for line in file:
        c_name, c_phone = line.strip().split(',')
        contact_data[c_name] = c_phone

  return contact_data

def main():
  filename = "contact.txt"
  contacts = load_contactdetails(filename)
  while True:
    print("\n\n---------Contact Management System-----------\n\n")

    print("Select: 1: To Add New Contacts Name and Phone Number\n")
    print("Select: 2: To Delete Existing Contact \n")
    print("Select: 3: To Update Old Contact Phone Number reference to Existing Contact name\n")
    print("Select: 4: To Update Old Contact Name reference to Existing Contact Phone Number\n")
    print("Select: 5: To View All contact Details from Database \n")
    print("Select: 6: To Exit The Application \n")

    choice = input("Select Your Choice: ")

    #adding new Contact

    if(choice == '1'):
      c_name = input("Enter Contact name: \n")
      c_phone = input("Enter Contact Number: \n")
      add_new_contact(contacts, c_phone, c_name)
      save_contact_file(contacts, filename)

    #deleting existing contact

    elif(choice == '2'):
      c_name = input("Enter Contact name to be deleted: \n")
      del_old_contact(contacts, c_name)
      save_contact_file(contacts, filename)

    #Updating existing contact phone number to new Number

    elif(choice == '3'):
      c_name = input("Enter the Old Contact name to be Updated: \n")
      new_phonenum = input("Enter The Contact New Phone Number: \n")
      update_old_contactphone(contacts, c_name, new_phonenum)
      save_contact_file(contacts, filename)

    #Updating existing contact name to new name

    elif(choice == '4'):
      c_new_name = input("Enter The new Contact name to update: \n")
      c_phone = input("Enter The existing phone number: \n")
      update_old_contactname(contacts, c_new_name, c_phone)
      save_contact_file(contacts, filename)

    #View all contacts details

    elif(choice == '5'):
      view_all_contactdetails(contacts)

    #Exit the application

    elif(choice == '6'):
      print("Goodbye!")
      break

    else:
      print("Invalid choice. Please try again...........\n")


if __name__ == "__main__":
    main()
```
